```python
import sympy as sp
import numpy as np 
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython import display
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import HTML
from scipy import special
```

```python
nx=511
nz=200
xf=0.006
xs=np.linspace(-xf,xf,nx)
xss=xs*1000
dx=np.diff(xs)[0]
zf=1.5
zs=np.linspace(0,zf,nz)
dz=np.diff(zs)[0]
X,Z=np.meshgrid(xss,zs)
```

$$\frac{\partial^2 U}{\partial x^2}+i2k\frac{\partial U}{\partial z}=0$$

reacomodando
$$-\frac{1}{i2k}\frac{\partial^2 U}{\partial x^2}=\frac{\partial U}{\partial z}$$

$$\therefore \kappa =-\frac{1}{i2k}$$

$$\alpha =\kappa \frac{\Delta z}{{\Delta x}^2}$$

$$k=\frac{2\pi}{\lambda}$$




```python
l=632.8*1e-9
k=(2*np.pi/l)
kappa=-1/(2*1j*k)

a=kappa*dz/(dx**2)
h=0.001

b=0.0002#0.2 mm

##C-inversa

v1=np.ones(nx)*2*(1+a)
v2=np.ones(nx-1)*(-a)
C0=np.diag(v1,k=0)
C1=np.diag(v2,k=1)
C2=np.diag(v2,k=-1)

C=np.linalg.inv((C1+C2+C0))
#D
v3=np.ones(nx)*2*(1-a)
D1=np.diag(v3,k=0)

D=(-C1-C2+D1)

P=np.matmul(C,D)

nxx=int(nx/2)
U=np.zeros((nz,nx))*1j
Ul=np.zeros(nx)
```

## Pulso inicial y condiciones de frontera

```python
[sn, cn, dn, phi] = special.ellipj(xs*1000, 0.7);
#plt.plot(xss,cn)

#po=np.real(np.exp(1j*((k)*xs)))#+np.real(np.exp(1j*((k+b)*xs)))#=np.cos(1*xs)#2/np.cosh(xs)#np.exp(-xs**2)#np.sin(3*xs)+
#po=np.exp((-(xs)**2)/0.01)
#po=np.exp((-(xs-h/2)**2)/0.0000001)+np.exp((-(xs+h/2)**2)/0.0000001)
#po=np.exp((-(xs)**2)/0.0000001)
#po=dn
po=1/np.cosh((xs-h/2)/0.0001)+1/np.cosh((xs+h/2)/0.0001)
#po=np.cos(k*xs)
#po=np.exp(-((xs)**2)/0.000001)*np.sin(k*xs)


U[:,0]=0
U[:,-1]=0

U[0,:]=po
#Barrera izquiera
U[0,0:nxx-int((h/(2*dx)).round(0))-int((b/(2*dx)).round(0))]=0

#barrera derecha
U[0,nxx+int((h/(2*dx)).round(0))+int((b/(2*dx)).round(0)):-1]=0
U[0,-1]=0

# Barrera Central
U[0,nxx-int(h/(2*dx))+int(b/(2*dx)):nxx]=0
U[0,nxx:nxx+int((h/(2*dx)).round(0))-int((b/(2*dx)).round(0))]=0
```

## Propagacion en z del pulso

```python
for i in range(0,nz-1):
    us=np.matmul(P,U[i,:])
    U[i+1,:]=us
```

```python
fig=plt.figure(figsize=(10,15))
ax = fig.add_subplot(111)
ax.clear()
prop=ax.contourf(X,Z,np.abs(U)**2,levels=200,cmap='jet')
plt.xlim([-6,6])
plt.colorbar(prop)
lwd=2.6
col='white'
#plt.xlim([-xf-2,xf+2])
ax.plot(xss[0:nxx-int(h/(2*dx))-int(b/(2*dx))],xss[0:nxx-int(h/(2*dx))-int(b/(2*dx))]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx+int(h/(2*dx))+int(b/(2*dx)):-1],xss[nxx+int(h/(2*dx))+int(b/(2*dx)):-1]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx-int(h/(2*dx))+int(b/(2*dx)):nxx],xss[nxx-int(h/(2*dx))+int(b/(2*dx)):nxx]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx:nxx+int(h/(2*dx))-int(b/(2*dx))],xss[nxx:nxx+int(h/(2*dx))-int(b/(2*dx))]*0,1,color=col,lw=lwd)
#plt.title('Propagacion abs(U)^2 en z')
plt.title(r'Propagacion $|U|^{2}$ en $z$', fontsize='large')
plt.xlabel(r'$X$ mm')
plt.ylabel(r'$Z$ meters')
```

```python
def df(y,dt,ti,tf):
    t=np.arange(ti,tf,dt)
    n=len(y)
    v1=np.ones(n-1)*(1/2)
    v2=np.ones(n-1)*(-1/2)
    D1=np.diag(v1,k=1)
    D2=np.diag(v2,k=-1)
    D=(D1+D2)/dt
    yp=np.matmul(D, y)
    yp[0]=yp[1]
    yp[-1]=yp[-2]
    ##Plot##############################################################
    #plt.plot(t,y,color='black',label='funcion')
    plt.plot(t,yp,color='red',label='Derivada numerica')
    #plt.axis('equal')
    plt.grid('on')
    plt.legend()
    plt.show()
    return yp
```

## 2. Propagacion en z del pulso con una lente

Veamos el efecto de colocar una lente convergente a mitad de la distancia de propagacin,
ver la Fig. 1(b). Para esto propaga el campo como se hizo en el Prob. 1 pero ahora hasta z = 0.75 m.
En ese lugar multiplica el campo por la transmitancia de una lente convergente


$$ T_L(x)= exp{(-i\frac{k}{2f}x^2)},$$
$$f=0.75 m$$

```python
f=0.75
Tl=np.exp((-1j*k*xs**2)/(2*f))
for i in range(0,nz-1):
    if zs[i].round(2)==f:
         us=np.matmul(P,U[i,:])*Tl
    else:
        us=np.matmul(P,U[i,:])
    U[i+1,:]=us
    
    
```

```python
fig3=plt.figure(figsize=(10,15))
ax = fig3.add_subplot(111)
prop=ax.contourf(X,Z,np.abs(U)**2,levels=200,cmap='jet')
plt.colorbar(prop)

ff=np.ones(nx)*zff
ax.plot(xss,ff,1,color='white',lw=3,label='Lente1')
plt.legend()

lwd=2.6
col='white'
#plt.xlim([-xf-2,xf+2])
ax.plot(xss[0:nxx-int(h/(2*dx))-int(b/(2*dx))],xss[0:nxx-int(h/(2*dx))-int(b/(2*dx))]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx+int(h/(2*dx))+int(b/(2*dx)):-1],xss[nxx+int(h/(2*dx))+int(b/(2*dx)):-1]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx-int(h/(2*dx))+int(b/(2*dx)):nxx],xss[nxx-int(h/(2*dx))+int(b/(2*dx)):nxx]*0,1,color=col,lw=lwd)
ax.plot(xss[nxx:nxx+int(h/(2*dx))-int(b/(2*dx))],xss[nxx:nxx+int(h/(2*dx))-int(b/(2*dx))]*0,1,color=col,lw=lwd)

plt.title('Propagacion en z del pulso con un lente  f='+str(f))
plt.xlabel('X mm')
plt.ylabel('Z meters')
```
