## Lente Convergente


```python
U=np.zeros((nx,ny,nt))
Tl=np.exp((-1j*k*Y**2)/(2*f))
f=8.6
for tt in range(1,nt-1):
    U[0,:,:]=0
    U[-1,:,:]=0
    U[:,0,:]=0
    U[:,-1,:]=0
    U[0:-1,int(ny/2),tt]=dt**2*20*np.sin(30*np.pi*ts[tt]/20)#np.cos(3-0.1*ts[tt])###########Onda
    #der
    #U[0:int(ny/2)-int(h/(2*dy))-int(b/(2*dy)),int(3*nx/4),:]=0
    #izq
    #U[int(ny/2)+int(h/(2*dy))+int(b/(2*dy)):-1,int(3*nx/4),:]=0
    #centr
    #U[int(ny/2)-int(h/(2*dy)):int(ny/2)+int(h/(2*dy)),int(3*nx/4),:]=0
    for i in range(1,nx-1):
        for j in range(1,ny-1):
            if ys[j].round(2)==f:
                U[i,j,tt+1]=(2*U[i,j,tt]-U[i,j,tt-1]+CFL**2*((U[i+1,j,tt])+U[i,j+1,tt]-4*U[i,j,tt]+U[i-1,j,tt]+U[i,j-1,tt]))*Tl
            else:
                U[i,j,tt+1]=2*U[i,j,tt]-U[i,j,tt-1]+CFL**2*((U[i+1,j,tt])+U[i,j+1,tt]-4*U[i,j,tt]+U[i-1,j,tt]+U[i,j-1,tt])
```
