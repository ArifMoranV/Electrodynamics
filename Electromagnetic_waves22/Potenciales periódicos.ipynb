```julia
using Plots
using LaTeXStrings
```

```julia
ħ=1/(2*π)
N=2000
NP=10
α=1/ħ
a=repeat(range(-2*π,2*π,length=N)',NP)'
b=1
k=1

#k=1#range(0,1000,length=1000)
```

```julia
P=repeat(range(0.1,10,length=NP)',N)
```

```julia
ϵ =(P.*sin.(α*a)./(α*a)).+ cos.(α*a)

plot(k*a,ϵ,label="",legendfont=font(10),xlabel=L"cos(ka/ka_0)",ylabel=L"\epsilon",grid="off",size = (800, 500),alpha=0.6,color = "black")

plot!([-2*π,2*pi],[1,1], linestyle=:dash, color=:red, label="1")
plot!([-2*π,2*pi],[-1,-1], linestyle=:dot, color=:blue, label="-1")
```
